```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/data/mcdonalds.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
data=df.iloc[:,1:11]
```

```python
data.head()
```

```python
li=['convenient','spicy','fattening','greasy','fast','cheap','tasty','expensive','healthy','disgusting']
```

```python
for i in range(len(li)):
  data[li[i]]=np.where(data[li[i]]=='Yes',1,0)

```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
scale=st.fit_transform(data)
```

```python
data.head()
```

```python
print(scale[:5])
```

```python
y=df['Gender']
x=pd.DataFrame(scale, columns=data.columns)
```

```python
x.head()
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn. decomposition import PCA
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(x)
```

```python
clusters=kmeans.predict(x)
clusters
```

```python
pca=PCA(n_components=2)
x_reduced=pd.DataFrame(pca.fit_transform(x),columns=["PC1","PC2"])
```

```python
x_reduced
```

```python
plt.figure(figsize=(14,10))
plt.scatter(x_reduced['PC1'],x_reduced['PC2'])
```

```python
reduced_centers=pca.transform(kmeans.cluster_centers_)
```

```python
reduced_centers
```

```python
x_reduced['cluster']=clusters
```

```python
plt.figure(figsize=(14,10))
plt.scatter(x_reduced[x_reduced['cluster']==0].loc[:,'PC1'],x_reduced[x_reduced['cluster']==0].loc[:,'PC2'],color='green')
plt.scatter(x_reduced[x_reduced['cluster']==1].loc[:,'PC1'],x_reduced[x_reduced['cluster']==1].loc[:,'PC2'],color='red')
plt.scatter(reduced_centers[:,0],reduced_centers[:,1],color='black',marker='x',s=300)
plt.show()
```
